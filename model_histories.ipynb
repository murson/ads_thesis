```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
autumn = np.load('./autumn_v2_resunet_512_model_history.npz')
spring = np.load('./spring_v2_resunet_512_model_history.npz')
combined_small = np.load('./combined_small_v2_resunet_512_model_history.npz')
combined_large = np.load('./combined_v2_resunet_512_model_history.npz')

models = [autumn, spring,combined_small,combined_large]

model_names = ['Autumn 2018','Spring 2018','Combined Small','Combined Large']


```

```python
def plot_history(model, model_name):
    fig, axs = plt.subplots(1, 2, sharey = True)
    axs[0].plot(model['mean_iou'])
    axs[0].plot(model['val_mean_iou'])
    axs[1].plot(model['loss'])
    axs[1].plot(model['val_loss'])
    axs[0].legend(['mean_iou', 'val_mean_iou'], loc='upper right')
    axs[1].legend(['loss', 'val_loss'], loc='upper right')
    axs[0].set_xlabel('Epochs')
    fig.suptitle(model_name)
    fig.savefig(model_name+'.png')
```

```python
for i,model in enumerate(models):
    plot_history(model, model_names[i])
```
